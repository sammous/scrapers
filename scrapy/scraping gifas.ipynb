```python
import requests
from lxml import etree

url = 'https://www.gifas.asso.fr/recherche_adherents?1'
xpath = '/html/body/div[7]/div[3]/div[1]/div[2]/div[2]/ul/li[2]/div[1]/form/div/div[3]/table/tbody/tr[1]/td[2]/select/option[2]'
```

```python
get = requests.get(url, stream=True)
get.raw.decode_content = True

l = get.content.split('\n')
l = map(lambda x: x.strip(), l)
entreprises = l[527:911]
```

```python
entreprises[16].encode('utf8')
```

```python
url_entreprises = []
url_entreprise = 'https://www.gifas.asso.fr/node/'
with open('gifas.txt', 'wb') as f:
    for i in entreprises:
        if '&' in i:
            i = i.replace('&', 'and')
        print(i)
        p = etree.fromstring(i)
        a = url_entreprise + str(p.attrib['value'])
        b = p.text.encode('utf8')
        f.write(b + ', ' + a + '\n')
        url_entreprises.append(a)
    f.close()
```
